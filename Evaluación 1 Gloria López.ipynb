```python
# Ejercicio 1: Clase TiendaOnline

class TiendaOnline:
    def __init__(self):
        self.inventario = {}  # Un diccionario para almacenar los productos del inventario: {nombre: [precio, cantidad]}
        self.clientes = {}  # Un diccionario para registrar a los clientes: {id_cliente: [email, compras[]]}
        self.ventas = 0.0  # ventas totales, inicializado con valor 0

# 1. Agregar productos al inventario:
    def agregar_producto(self, id_producto, nombre, precio, cantidad):
        # Verificar si el id_producto no existe en el diccionario de productos
        if id_producto not in self.inventario:
            self.inventario[id_producto] = {
                "id": id_producto,
                "nombre": nombre,
                "precio": precio,
                "cantidad": cantidad
            }
            # Imprimir un mensaje de éxito con el nombre del producto:
            print(f"Producto {nombre} agregado con éxito.")
            # Devolver el diccionario de productos actualizado:
            return self.inventario
        else:
            # Si el producto ya existe, muestra mensaje de error
            print(f"El producto con ID {id_producto} ya existe en el inventario")

# 2. Ver inventario:
    def ver_inventario(self):
        for id_producto, detalles in self.inventario.items():
            print(f"ID: {id_producto}, Nombre: {detalles['nombre']}, Precio: {detalles['precio']}, Cantidad: {detalles['cantidad']}")
            
            
# 3. Buscamos el producto utilizando bucle for:
    def buscar_producto(self, nombre_buscar):
        # Recorro el inventario. Busco el nombre en minúsculas:
        nombre_buscar = str(nombre_buscar).lower()
        for id_producto, detalles in self.inventario.items():
            # si el nombre está, muestra los detalles de precio y cantidad:
            if detalles["nombre"].lower() == nombre_buscar:
                precio = detalles["precio"]
                cantidad = detalles["cantidad"]
                # Imprimir mensaje con el nombre del artículo (la primera en mayúsculas), precio y cantidad y paro con break:
                print(f"Nombre: {detalles['nombre'].capitalize()}, Precio: ${precio}, Cantidad: {cantidad}")
                break
        else:
            # Si el nombre no está, imprimir que el producto (primera en mayúsculas) no está en el inventario:
            print(f"El producto '{nombre_buscar.capitalize()}' no se encontró en el inventario.")
            
# 4. Actualizamos stock de un producto del inventario:
    def actualizar_stock(self, nombre, cantidad):
        # recorro el inventario: busco el nombre en minúsculas
        nombre = nombre.lower()
        for detalles in self.inventario.values():
            if detalles["nombre"].lower() == nombre:
        # se actualiza la cantidad y se imprime un mensaje de que el producto ha sido actualidado con los detalles:
                detalles["cantidad"] += cantidad
                print(f"Stock de '{detalles['nombre'].capitalize()}' actualizado a {detalles['cantidad']}")
                break
        # si no está el nombre, se imprime un mensaje que no está en el inventario:
        else:
            print(f"El producto '{nombre.capitalize()}' no está en el inventario.")
            
# 5.  Eliminar producto del inventario por nombre:
    def eliminar_producto(self, nombre):
        # recorro el inventario: busco el nombre en minúsculas
        nombre = nombre.lower()
        for id_producto, detalles in self.inventario.items():
        # si el producto buscado está en el inventario, lo elimino 
            if detalles["nombre"].lower() == nombre:
                del self.inventario[id_producto]
        # se imprime que el nombre (con la priemra en mayúsculas) ha sido eliminado del inventario y se termina
                print(f"Producto '{detalles['nombre'].capitalize()}' eliminado del inventario.")
                break
        # se actualiza que ese producto ya no está en el inventario
        print(f"El producto '{nombre.capitalize()}' no está en el inventario.")

# 6. Calcular el valor total del inventario:
    def calcular_valor_inventario(self):
        desglose = {}  # Diccionario para almacenar el valor de cada producto
        valor_total = 0  # Variable para almacenar el valor total del inventario
        # para los detalles que son valores en el inventario:
        for detalles in self.inventario.values():
            precio = detalles["precio"]
            cantidad = detalles["cantidad"]
            # el valor del producto = precio por la cantidad
            valor_producto = precio * cantidad
            desglose[detalles["nombre"]] = valor_producto
            valor_total += valor_producto

        return desglose, valor_total

#7. Realizar una compra seleccionando productos del inventario:
    def realizar_compra(self):
        # creo diccionario para carrito
        continuar_comprando = True
        carrito = {}
        # Bucle while que se ejecutará mientras la variable continuar_comprando sea verdadera. Esto permite que el cliente siga agregando productos a su carrito hasta que decida finalizar la compra.
        while continuar_comprando:
            # se imprime un mensaje para informar al cliente del invetario disponible
            print("Inventario disponible:")
            self.ver_inventario()
            # el cliente elige el nombre del producto que quiere comprar. si quiere terminar la compra, pone fin
            nombre_producto = input("Ingrese el nombre del producto que desea comprar (o 'fin' para finalizar la compra): ")
            if nombre_producto.lower() == "fin":
                continuar_comprando = False
            else:
                producto_encontrado = False
                # se itera en el invetario para buscar el producto que ha elegido el cliente y si está, se sigue con el proceso de compra:
                for id_producto, detalles in self.inventario.items():
                    if detalles["nombre"].lower() == nombre_producto.lower():
                        producto_encontrado = True
                        cantidad_disponible = detalles["cantidad"]
                        # se selecciona la cantidad a comprar:
                        cantidad_a_comprar = int(input(f"Ingrese la cantidad que desea comprar de '{detalles['nombre'].capitalize()}' (disponibles: {cantidad_disponible}): "))
                        if cantidad_a_comprar <= cantidad_disponible:
                            if id_producto in carrito:
                                carrito[id_producto] += cantidad_a_comprar
                            else:
                            # si la cantidad que se quiere comprar es válida y está en el inventario, se añade al carrito
                                carrito[id_producto] = cantidad_a_comprar
                            self.inventario[id_producto]["cantidad"] -= cantidad_a_comprar
                            print(f"'{detalles['nombre'].capitalize()}' se ha añadido al carrito.")
                        else:
                            print("La cantidad solicitada excede la cantidad disponible en el inventario.")
                        break
                if not producto_encontrado:
                    print("El producto ingresado no existe en el inventario.")
        # se calcula el coste total de la compra
        coste_total = 0
        for id_producto, cantidad in carrito.items():
            precio = self.inventario[id_producto]["precio"]
            coste_total += cantidad * precio
        # se finaliza la compra:
        return carrito, coste_total

    # 8. Procesar pago:
    def procesar_pago(self, coste_total):
        try:
            # se solicita la cantidad para pagar
            cantidad_pagada = float(input("Ingrese la cantidad de pago: $"))
            # si la cantidad pagada es menor al coste total, se imprime el error que es insuficiente:
            if cantidad_pagada < coste_total:
                raise ValueError("El monto pagado es insuficiente.")
            # el cambio es la cantidad introducida menos el coste total, con dos decimales
            cambio = cantidad_pagada - coste_total
            print(f"Pago exitoso. Su cambio es: ${cambio:.2f}")
        # si el cliente ingresa una cantidad que no sea un número, se impirmie un mensaje diciendo que es un error
        except ValueError as e:
            print(f"Error: {e}")

# 9. Agregar cliente con correo electrónico:
    def agregar_cliente(self, nombre, email):
        # verificar si el cliente está registrado en el diccionario de clientes:
        if nombre not in self.clientes:
            # si el cliente no está en el registro, se agrega al diccionario de clientes, el nombre y el email.
            self.clientes[nombre] = {'email': email, 'compras': []}
            # 
            print(f"Cliente {nombre} agregado con éxito.")
        else:
            print("El cliente ya existe en el registro.")

        
# 10.Ver clientes utilizando bucle for:
    def ver_clientes(self):
        self.clientes[nombre] = email
        for nombre,email in ver_clientes.items():
            print(f"Nombre: {nombre}, Email: {email}")
            
# 11. Registrar compra para un cliente. Actualizar ventas totales y agregar la compra al historial del cliente.
    def registrar_compra(self,nombre_cliente,carrito):
        # para registrar la compra, hay que ver si el nombre del cliente está en el diccionario de clientes. Si no está, se imprime un documento diciendo que no se puede completar la acción:
        if nombre_cliente not in self.clientes:
            print(f"No se puede realizar la acción. El cliente {nombre_cliente} no está en el diccionario de clientes.")
            return
        # el total de la compra esigual a la suma del precio por la cantidad de los artículos de carrito. La compra se añade al carrito del cliente
        total_compra = sum(detalles['precio'] * detalles['cantidad'] for detalles in carrito.values())
        self.clientes[nombre_cliente]['compras'].append({'carrito': carrito, 'total': total_compra})
        self.ventas += total_compra
        print(f"Compra registrada para {nombre_cliente}. Total: ${total_compra:.2f}")

# 12. Ver_compras_cliente(self, nombre_cliente)`: Muestra el historial de compras de un cliente. Debe recibir el nombre del cliente como parámetro.
        # Busca al cliente en el diccionario de clientes.
        # Muestra las compras realizadas por el cliente, incluyendo detalles de productos y totales.
    def ver_compras_cliente(self, nombre_cliente):
        if nombre_cliente in self.clientes:
            historial_compras = self.clientes[nombre_cliente]['compras']
            if historial_compras:
                print(f"Historial de compras para el cliente {nombre_cliente}:")
                for compra in historial_compras:
                    carrito = compra['carrito']
                    total_compra = compra['total']
                    print(f"Compra:")
                for producto, detalles in carrito.items():
                    print(f"   - Producto: {producto.capitalize()}, Precio: ${detalles['precio']}, Cantidad: {detalles['cantidad']}")
                print(f"   Total: ${total_compra:.2f}")
            else:
                print(f"No hay compras registradas para el cliente {nombre_cliente}.")
        else:
            print(f"El cliente {nombre_cliente} no está registrado.")
            
# 13. Calcular_ventas_totales(self)`: Muestra las ventas totales de la tienda.Suma los totales de todas las compras realizadas y muestra el total de las ventas totales en la tienda.
    def calcular_ventas_totales(self):
        ventas_totales = sum(compra['total'] for cliente in mi_tienda.clientes.values() for compra in cliente.get('compras', []))
        self.ventas_totales = ventas_totales
        print(f"Las ventas totales de la tienda son: ${ventas_totales:.2f}")
```
